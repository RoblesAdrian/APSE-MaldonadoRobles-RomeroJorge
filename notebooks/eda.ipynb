```python
items = pd.DataFrame(df['items'].explode())
items = items['items'].apply(pd.Series)

trucks = pd.DataFrame(df['trucks'].explode())
trucks = trucks['trucks'].apply(pd.Series)
```

```python
items = pd.DataFrame(items)
trucks = pd.DataFrame(trucks)
```
